```python
# Phase 1: Inspect dad_df structure
print("dad_df shape:", dad_df.shape)
print("\ndad_df columns:")
print(dad_df.columns.tolist())
dad_df[['PATID', 'ADMITDATE_DT', 'DISDATE_DT', 'episode_order']].head(10)
```

```python
# Phase 2: Check episode_order uniqueness in dad_df and ed_df
for name, df in [('dad_df', dad_df), ('ed_df', ed_df)]:
    total = len(df)
    unique_ep = df['episode_order'].nunique()
    print(f"{name}:")
    print(f"  Total rows:              {total:>10,}")
    print(f"  Unique episode_orders:   {unique_ep:>10,}")
    print(f"  Duplicate rows:          {total - unique_ep:>10,}")
    print(f"  Is unique: {df['episode_order'].is_unique}")
    print()
```

```python
# Phase 3: Main merge — ecg + dad (inpatient) + ed (emergency)
dad_slim = dad_df[['episode_order', 'ADMITDATE_DT', 'DISDATE_DT']].copy()
ed_slim  = ed_df[['episode_order', 'VISIT_DATE_DT']].copy()
ecg_base = ecg_df[['patient_id', 'ecg_id', 'ecg_acquisition_time', 'episode_id']].copy()

print(f"ecg_base:  {ecg_base.shape}")
print(f"  ECGs with episode_id:    {ecg_base['episode_id'].notna().sum():>10,}")
print(f"  ECGs without episode_id: {ecg_base['episode_id'].isna().sum():>10,}")

# Left join with DAD (inpatient: ADMITDATE_DT, DISDATE_DT)
merged = ecg_base.merge(
    dad_slim,
    left_on='episode_id',
    right_on='episode_order',
    how='left'
).drop(columns='episode_order')
print(f"\nAfter DAD merge:  {merged.shape}  (fan-out: +{len(merged) - len(ecg_base):,} rows)")

# Left join with ED (emergency: VISIT_DATE_DT)
merged = merged.merge(
    ed_slim,
    left_on='episode_id',
    right_on='episode_order',
    how='left'
).drop(columns='episode_order')
print(f"After ED  merge:  {merged.shape}  (fan-out: +{len(merged) - len(ecg_base):,} rows total)")
```

```python
# Phase 4: Deduplication — keep row nearest in time to ecg_acquisition_time
n_before = len(merged)
if n_before > len(ecg_base):
    ecg_t    = pd.to_datetime(merged['ecg_acquisition_time'])
    admit_dt = pd.to_datetime(merged['ADMITDATE_DT'], errors='coerce')
    visit_dt = pd.to_datetime(merged['VISIT_DATE_DT'], errors='coerce')

    # Time diff to each available date; take whichever is smaller
    dad_diff = (admit_dt - ecg_t).abs()
    ed_diff  = (visit_dt - ecg_t).abs()
    merged['_time_diff'] = pd.concat([dad_diff, ed_diff], axis=1).min(axis=1, skipna=True)
    merged['_time_diff'] = merged['_time_diff'].fillna(pd.Timedelta(0))  # no episode rows

    merged = (merged
              .sort_values(['ecg_id', '_time_diff'])
              .drop_duplicates(subset='ecg_id', keep='first')
              .drop(columns='_time_diff')
              .reset_index(drop=True))

    print(f"Rows before dedup: {n_before:,}")
    print(f"Rows after  dedup: {len(merged):,}  (removed {n_before - len(merged):,} duplicate rows)")
else:
    merged = merged.reset_index(drop=True)
    print("No fan-out detected — deduplication not needed.")
```

```python
# Phase 5: Final AHS ECG-Episode linkage table
ahs_linkage = merged[['patient_id', 'ecg_id', 'ecg_acquisition_time', 'episode_id',
                       'ADMITDATE_DT', 'DISDATE_DT', 'VISIT_DATE_DT']].copy()

print("=" * 55)
print("  AHS ECG–Episode Linkage Table")
print("=" * 55)
print(f"  Total ECGs                    : {len(ahs_linkage):>10,}")
print(f"  ECGs with any episode_id      : {ahs_linkage['episode_id'].notna().sum():>10,}")
print(f"  Linked to DAD (inpatient)     : {ahs_linkage['ADMITDATE_DT'].notna().sum():>10,}")
print(f"  Linked to ED                  : {ahs_linkage['VISIT_DATE_DT'].notna().sum():>10,}")
print(f"  Linked to both DAD + ED       : {(ahs_linkage['ADMITDATE_DT'].notna() & ahs_linkage['VISIT_DATE_DT'].notna()).sum():>10,}")
print(f"  No episode (unlinked)         : {ahs_linkage['episode_id'].isna().sum():>10,}")
print("=" * 55)

ahs_linkage
```
